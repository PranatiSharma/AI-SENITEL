```python
with open("clean_cryptojacking_signatures.xlsx", "r", encoding="utf-8", errors="ignore") as f:
    lines = f.readlines()

print(lines[:20])


```

```python
signatures = []

for line in lines:
    line = line.strip()
    if not line:
        continue
    if line.startswith("!") or line.startswith("["):
        continue
    signatures.append(line)

print("Total signatures loaded:", len(signatures))
print(signatures[:10])
```

```python
crypto_domains = set()
crypto_patterns = []

for sig in signatures:
    # Domain rule
    if sig.startswith("||"):
        domain = sig.replace("||", "").split("^")[0]
        crypto_domains.add(domain)

    # Regex / script pattern
    elif sig.startswith("/") and sig.endswith("/"):
        crypto_patterns.append(sig.strip("/"))

    else:
        crypto_patterns.append(sig)

print("Domains:", len(crypto_domains))
print("Patterns:", len(crypto_patterns))
```

```python
from urllib.parse import urlparse

def scan_scripts(soup):
    alerts = []

    for script in soup.find_all("script"):
        # External scripts
        if script.get("src"):
            src = script["src"]
            domain = urlparse(src).netloc

            if domain in crypto_domains:
                alerts.append(f"Cryptomining domain detected: {domain}")

            for pattern in crypto_patterns:
                if pattern.lower() in src.lower():
                    alerts.append("Cryptomining script pattern detected")

        # Inline scripts
        if script.string:
            content = script.string.lower()
            for pattern in crypto_patterns:
                if pattern.lower() in content:
                    alerts.append("Inline cryptomining code detected")

    return alerts
```

```python
import requests
from bs4 import BeautifulSoup

def analyze_website(url):
    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()  # Raise an HTTPError for bad responses (4xx or 5xx)
    except requests.exceptions.RequestException as e:
        print(f"Error fetching {url}: {e}")
        return

    soup = BeautifulSoup(response.text, 'html.parser')
    alerts = scan_scripts(soup)

    if alerts:
        print(f"--- Analysis for {url} ---")
        for alert in alerts:
            print(f"  - {alert}")
    else:
        print(f"No cryptomining signatures detected on {url}")

analyze_website("https://example.com")
```

```python
# ===============================
# SIGNATURE-BASED WEB DETECTOR
# ===============================

import requests
from bs4 import BeautifulSoup
from urllib.parse import urlparse

# --------------------------------
# 1. LOAD CRYPTOJACKING SIGNATURES
# --------------------------------

crypto_domains = set()
crypto_patterns = []

# Load NoCoin-style cryptojacking list
with open("clean_cryptojacking_signatures.csv", "r", encoding="utf-8", errors="ignore") as f:
    for line in f:
        line = line.strip()

        if not line or line.startswith("!") or line.startswith("["):
            continue

        # Domain rules
        if line.startswith("||"):
            domain = line[2:].split("^")[0]
            crypto_domains.add(domain)

        # Script patterns
        else:
            crypto_patterns.append(line.lower())

# --------------------------------
# 2. LOAD URLHAUS MALICIOUS DOMAINS
# --------------------------------

urlhaus_domains = set()
urlhaus_url = "https://urlhaus.abuse.ch/downloads/csv/"
r = requests.get(urlhaus_url, timeout=30)

with open("urlhaus.csv", "wb") as f:
    f.write(r.content)

with open("urlhaus.csv", "r", encoding="latin1", errors="ignore") as f:
    for line in f:
        if line.startswith("#"):
            continue
        parts = line.split(",")
        if len(parts) > 2:
            url = parts[2].strip().strip('"')
            if url.startswith("http"):
                domain = urlparse(url).netloc
                if domain:
                    urlhaus_domains.add(domain)

# --------------------------------
# 3. ADD BASELINE CRYPTOMINING SIGS
# --------------------------------

baseline_domains = {
    "coinhive.com",
    "webminerpool.com",
    "cryptoloot.pro"
}

baseline_patterns = [
    "coinhive",
    "cryptonight",
    "hashrate",
    "miner.start",
    "webassembly.instantiate"
]

crypto_domains.update(baseline_domains)
crypto_patterns.extend(baseline_patterns)

# Combine all malicious domains
malicious_domains = crypto_domains.union(urlhaus_domains)

# --------------------------------
# 4. DOMAIN NORMALIZATION FUNCTION
# --------------------------------

def is_malicious_domain(domain):
    return any(
        domain == d or domain.endswith("." + d)
        for d in malicious_domains
    )

# --------------------------------
# 5. SCRIPT SCANNER (SIGNATURE-BASED)
# --------------------------------

def scan_scripts(soup):
    alerts = []

    for script in soup.find_all("script"):
        # External scripts
        if script.get("src"):
            src = script["src"]
            domain = urlparse(src).netloc

            if is_malicious_domain(domain):
                alerts.append(f"Malicious script domain: {domain}")

            for pattern in crypto_patterns:
                if pattern in src.lower():
                    alerts.append("Cryptomining script pattern detected")

        # Inline scripts
        if script.string:
            content = script.string.lower()
            for pattern in crypto_patterns:
                if pattern in content:
                    alerts.append("Inline cryptomining code detected")

    return alerts

# --------------------------------
# 6. COOKIE SCANNER (SIGNATURE-BASED)
# --------------------------------

cookie_signatures = [
    "miner", "coinhive", "cryptonight",
    "hashrate", "session", "uid", "fingerprint"
]

def scan_cookies(response):
    alerts = []

    for cookie in response.cookies:
        combined = f"{cookie.name}={cookie.value}".lower()
        for sig in cookie_signatures:
            if sig in combined:
                alerts.append(f"Suspicious cookie detected: {cookie.name}")

    return alerts

# --------------------------------
# 7. FINAL ANALYZER (DECISION ENGINE)
# --------------------------------

def analyze_website(url):
    response = requests.get(url, timeout=10)
    soup = BeautifulSoup(response.text, "html.parser")

    script_alerts = scan_scripts(soup)
    cookie_alerts = scan_cookies(response)

    print("\n--- SCRIPT SCAN RESULTS ---")
    for a in script_alerts:
        print("ALERT:", a)

    print("\n--- COOKIE SCAN RESULTS ---")
    for a in cookie_alerts:
        print("ALERT:", a)

    if script_alerts or cookie_alerts:
        print("\nFINAL VERDICT: ❌ HARMFUL WEBSITE")
    else:
        print("\nFINAL VERDICT: ✅ SAFE WEBSITE")

# --------------------------------
# 8. SAFE TEST CASE (FOR DEMO)
# --------------------------------

def test_detection():
    test_html = """
    <html>
    <head>
    <script src="https://coinhive.com/lib.js"></script>
    <script>
        var miner = "cryptonight";
        console.log("hashrate started");
    </script>
    </head>
    </html>
    """
    soup = BeautifulSoup(test_html, "html.parser")
    return scan_scripts(soup)

# --------------------------------
# RUN TEST
# --------------------------------

print(test_detection())
```
